```python
import csv
import pandas as pd
import ast
import numpy as np
```

```python
main_file_headers = ["genome_id", "protein_id", "protein_seq", "taxonomic_label", "host_id"]


# Read both csv to be merged
df1_iterator = pd.read_csv('/project/jespinoza_1537/GENOMESDB/bacteriaDB/embeddings.csv', chunksize=5000,header=None)
df2_iterator = pd.read_csv('/project/jespinoza_1537/GENOMESDB/bacteriaDB/bacteriaRandSub.csv', chunksize=5000,header=None, names=main_file_headers)



is_first_chunk = True 

for chunk1, chunk2 in zip(df1_iterator, df2_iterator):
    
    # Create copy of chunk 2 and add the embedding to it at the end
    merged_chunk = chunk2.copy()

    # For every row in out.csv, convert to a list and make it a string
    vector_descriptions = chunk1.apply(lambda row: str(row.tolist()), axis=1)

    # Did not use join() since can get messy in a loop
    merged_chunk['embedding'] = vector_descriptions.values

    # Append to mergedTables.csv
    if is_first_chunk:
            # For the first chunk, write in 'w' mode (overwrite) and include the header.
            merged_chunk.to_csv("mergedTables.csv", mode='w', header=True, index=False)
            is_first_chunk = False
    else:
            # For subsequent chunks, append in 'a' mode and omit the header.
            merged_chunk.to_csv("mergedTables.csv", mode='a', header=False, index=False)

    # Updates if first_chunk so header is not written more then once
    is_first_chunk = False

    print(f"Processed and wrote a chunk of {len(chunk1)} rows.")


print(f"Successfully merged all data into mergedTables.csv")
```

### Data Aggregation

```python
filename = "filteredData90/mergedTables90.csv"
grouping_column = "genome_id"


total_sums = pd.DataFrame()
total_counts = pd.DataFrame()

df_iterator = pd.read_csv(filename, chunksize=5000, header=0)

print(f"Processing '{filename}' ")

# Stores the running totals for each genome_id 
# Each genome_id maps to a list with [0] = numpy array sum, [1] = instance count
embedding_aggregates = {}

for chunk in df_iterator:

    # converts 'embedding' column back into a vector
    embeddings = chunk['embedding'].apply(ast.literal_eval)

    
    for genome_id, vector in zip(chunk[grouping_column], embeddings):

        # Converts the embedding vector into a numpy array for faster addition
        vector_np = np.array(vector, dtype=float)

        if genome_id in embedding_aggregates:
                # If genome_id exists, add to the existing vector and increment count
                embedding_aggregates[genome_id][0] += vector_np
                embedding_aggregates[genome_id][1] += 1
        else:
                # If it's a new genome_id, initialize it
                embedding_aggregates[genome_id] = [vector_np, 1]

# Calculate the final averages and store in the final means table
final_means = {}
for genome_id, (sum_vector, count) in embedding_aggregates.items():
    final_means[genome_id] = sum_vector / count

# Convert this dictionary to a dataframe for viewing
mean_embeddings_df = pd.DataFrame.from_dict(final_means)

df_transposed = mean_embeddings_df.T
df_renamed = df_transposed.rename_axis('genome_id')

df_renamed.to_csv('meanEmbeddings90.csv',mode='w', index=True)

print("\n--- Final Mean Values per GENOME ---")
print(mean_embeddings_df)
```

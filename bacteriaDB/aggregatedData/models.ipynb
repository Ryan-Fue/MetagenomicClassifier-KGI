# Split Data for Training and Testing

```python
microbe_predictors = merged_df.iloc[:, 0:1280] 
microbe_type = merged_df['1281']
#microbe_scaled = (microbe_predictors - microbe_predictors.mean()) / microbe_predictors.std()
#microbe_scaled
```

```python
X_train = pd.read_csv('X_train.csv').drop(columns=['Unnamed: 0'])
Y_train = pd.read_csv('Y_train.csv').drop(columns=['Unnamed: 0'])
X_test = pd.read_csv('X_test.csv').drop(columns=['Unnamed: 0'])
Y_test = pd.read_csv('Y_test.csv').drop(columns=['Unnamed: 0'])
```

```python
X_train
```

```python
"""
!!!!!!!!!!!!!!!!!!!!!!!!
!!!!!!!!!!!!!!!!!!!!!!!!

DONT RUN THIS!!!!!!!!!!!!

!!!!!!!!!!!!!!!!!!!!!!!!!!
!!!!!!!!!!!!!!!!!!!!!!!!!!
"""


# Split data for training and testing

X = microbe_predictors
Y = microbe_type

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 42)

# Store sets to csv 
X_train.to_csv('X_train80.csv',mode='w', index=True)
Y_train.to_csv('Y_train80.csv',mode='w', index=True)
X_test.to_csv('X_test80.csv',mode='w', index=True)
Y_test.to_csv('Y_test80.csv',mode='w', index=True)

Y_train
```

# Train KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors = 5)
dir()
```

```python
#dataset = list(X_train.values())
#labels = list(Y_train.values())
KNN.fit(X_train, Y_train.values.ravel())
```

```python
from sklearn import metrics
import matplotlib
import matplotlib as mpl

Y_pred = KNN.predict(X_test)

print(metrics.accuracy_score(Y_test, Y_pred))
#print(metrics.f1_score(y_test, y_pred))
#print(metrics.recall_score(y_test, y_pred))


# Heatmap Code
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)

# Heat map
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',xticklabels=['virus','bacteria'], yticklabels=['virus','bacteria'])
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('K Nearest Neighbors CM')
plt.tight_layout()
plt.show()
```

```python
import umap.umap_ as umap

# Generate table for UMAP (No labels yet)
umap_out = umap.UMAP(n_components=2, random_state=42).fit_transform(X_test)
umap_out = pd.DataFrame(data=umap_out, columns=['umap1', 'umap2'], index=X_test.index)

umap_out['Predicted Microbe'] = Y_pred
umap_out['Y_pred'] = Y_pred
umap_out['Y_test'] = Y_test

umap_out.loc[(umap_out['Y_pred']== 1) & (umap_out['Y_test'] ==1), 'Predicted Microbe'] = 'TrueBact'
umap_out.loc[(umap_out['Y_pred']== 1) & (umap_out['Y_test'] ==0), 'Predicted Microbe'] = 'FalseBact'
umap_out.loc[(umap_out['Y_pred']== 0) & (umap_out['Y_test'] ==0), 'Predicted Microbe'] = 'TrueVirus'
umap_out.loc[(umap_out['Y_pred']== 0) & (umap_out['Y_test'] ==1), 'Predicted Microbe'] = 'FalseVirus'

umap_out.loc[(umap_out['Y_pred']== 0) & (umap_out['Y_test'] ==1), 'Predicted Microbe'] = 'FalseVirus'

sns.scatterplot(data=umap_out, x='umap1', y='umap2',hue = 'Predicted Microbe', palette='Set1')

plt.title("K Nearest Neighbors Classification")

plt.show()
```

```python

# Transpose df so Genome_ids are rows
X= df

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=11, random_state=42)
clusters = kmeans.fit(X)
kmeans.inertia_

cluster_groupings = kmeans.fit_predict(X)
```

```python
import umap.umap_ as umap

umap_out = umap.UMAP(n_components=2, random_state=42).fit_transform(X)

umap_out = pd.DataFrame(data=umap_out, columns=['umap1', 'umap2'], index=X.index)

umap_out['cluster'] = cluster_groupings
umap_out

#sns.scatterplot(data=umap_out, x='umap1', y='umap2',hue = 'cluster', palette='Set2')
```

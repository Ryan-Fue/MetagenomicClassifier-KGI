```python
bact_to_search = []
virus_to_search = []

# Extract list of indices of original table (merged_df) from X_test column 'Unnamed: 0'
indices = X_test.loc[:,'Unnamed: 0']

for index in indices:
    
    # Gets label (in column 1281) of item at index 
    label = merged_df.loc[index, '1281']
    
    if (label == 1):
        # Append the genome_id stored in column 0 of merged df for each index
        bact_to_search.append(merged_df.loc[index,'0'])
    elif (label == 0):
        virus_to_search.append(merged_df.loc[index,'0'])

# For faster look up
bact_to_search = set(bact_to_search)
virus_to_search = set(virus_to_search)
    
```

```python
df1_iterator = pd.read_csv('mergedTables.csv', chunksize=5000)

first_read = True
count = 0

# Read in chunks to save memory
for chunk in df1_iterator:
    
    
    for index, row in chunk.iterrows():

        # Check if the embedding is in X_test
        if row['genome_id'] in bact_to_search:

            # Convert embedding string into a list
            embedding_data = ast.literal_eval(row['embedding'])
            
            temp_df = pd.DataFrame([embedding_data])

            # Append a label to the list
            temp_df['genome_id'] = row['genome_id']
            temp_df['label'] = 0

            if first_read == True:
                temp_df.to_csv("singleTestData.csv", mode='w', header=False, index=False)
                first_read = False
            
            else:
                temp_df.to_csv("singleTestData.csv", mode='a', header=False, index=False)
    count+=1
    print(f"Chunk {count} was appended")

```

```python

# Virus filtering

virus_df = pd.DataFrame()
virus_rows_list = []

# Read csv by chunk to save RAM
df2_iterator = pd.read_csv('/project/jespinoza_1537/GENOMESDB/virusDB/www.genome.jp/80_clustered_data.csv', nrows=159740, chunksize = 5000)
for chunk2 in df2_iterator:

    # Iterate through table and append 
    for index, row in chunk2.iterrows():
        if row['Genome_ID'] in virus_to_search:

            # Append just the embeddings of the viruses
            virus_rows_list.append(row[3:])
    print("Chunk Completed")

virus_df = pd.DataFrame(virus_rows_list)

virus_df['label'] = 0

print("Virus Data Filtered")


virus_df.to_csv("singleTestData80.csv", mode='w', header=False, index=False)

print('DONE!!')
```

```python
# Bacteria filtering

df1_iterator = pd.read_csv('filteredData80/mergedTables80.csv', chunksize=5000)

bact_df = pd.DataFrame()
bact_rows_list = []
    
for chunk1 in df1_iterator:
    for index, row in chunk1.iterrows():
    
        # Check if the embedding is in X_test
        if row['genome_id'] in bact_to_search:
    
            # Convert embedding string into a list
            embedding_data = ast.literal_eval(row['embedding'])
            
            bact_row_df = pd.DataFrame([embedding_data])
    
            bact_row_df['genome_id'] = row['genome_id']
            
            # Combine with current table
            bact_rows_list.append(bact_row_df)
    print("Chunk Completed")
        
bact_df = pd.concat(bact_rows_list, ignore_index=True)

bact_df['label'] = 1

print("Bacteria Data Filtered")


bact_df.to_csv("singleTestData80.csv",mode='a', header=False, index=False)

print('DONE!!')
```
